```python
import wbdata
import pandas as pd
countries = ["AFR","CHN","USA","DNK","FRA","LBN","LKA","LUX"]
import datetime
df  = wbdata.get_dataframe({'NY.GDP.MKTP.KD.ZG': 'GDP Rate'}, country=countries)
df = df.unstack("country")
df.columns = df.columns.droplevel()
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.violinplot(data=df)
plt.title("GDP Growth Rate")
plt.show()
```

```python
sns.violinplot(data=df[(df<20) & (df>-10)])
plt.title("GDP Growth Rate")
plt.show()
```

```python
sns.violinplot(data=df[(df<20) & (df>-10)],inner="quart")
plt.title("GDP Growth Rate")
plt.show()
```

```python
sns.violinplot(data=df[(df<20) & (df>-10)],inner=None)
sns.swarmplot(data=df[(df<20) & (df>-10)],color='k')
plt.title("GDP Growth Rate")
plt.show()
```

```python
df.reset_index(inplace=True)
df
```

```python
df = pd.melt(df,id_vars=["date"],value_name="GDP Rate")
df
```

```python
print(df["date"])
```

```python
df["date"] = pd.to_numeric(df["date"])
```

```python
def name(x):
    if x<1980:
        return "Before"
    else:
        return "After"
df["Era"] = df["date"].apply(name)
df
```

```python
sns.violinplot(x="country", y="GDP Rate", hue="Era", data=df)
plt.show()
```

```python
sns.violinplot(x="Era", y="GDP Rate", hue="country", data=df)
plt.show()
```

```python
sns.violinplot(x="country", y="GDP Rate", hue="Era", data=df, split=True)
plt.show()
```

```python
sns.violinplot(x="country", y="GDP Rate", hue="Era", data=df, split=True,palette={"Before": "r", "After": "g"})
plt.show()
```

```python
sns.violinplot(x="country", y="GDP Rate", hue="Era", data=df[(df<20) & (df>-10)], split=True,palette={"Before": "r", "After": "g"})
plt.show()
```

```python
df = df.groupby("date").mean().reset_index()
df
```

```python
sns.factorplot(x="date", y="GDP Rate",data=df,kind="bar")
plt.show()
```

```python
graph = sns.factorplot(x="date", y="GDP Rate",data=df.groupby("date").mean().reset_index(),kind="bar")
graph.set_xticklabels(step=8)
plt.show()
```
