```python
#Let's create a function to find the total cost based on a fixed cost and a variable cost
def totalCost(fixedCost,variableCost,volume):
    return fixedCost+variableCost*volume
```

```python
#Let's get x and y values
xValues = list(range(0,101))
yValues = [totalCost(50,1,x) for x in xValues]
print(xValues)
print()
print(yValues)
```

```python
import matplotlib.pyplot as plt
plt.plot(xValues,yValues)
#Plot our x and y values
plt.show()
#Show the plot
```

```python
plt.plot(xValues,yValues)
plt.xlabel("Units")
#We can add an x label with xlabel()
plt.ylabel("Total Cost")
#And a y label with ylabel()
plt.title("Total Cost")
#Finally, a title
plt.show()
```

```python
def revenue(price,volume):
    return price*volume
```

```python
xValues = list(range(0,101))
yValues1 = [totalCost(50,1,x) for x in xValues]
yValues2 = [revenue(2,x) for x in xValues]
```

```python
plt.plot(xValues,yValues1)
plt.plot(xValues,yValues2)
plt.xlabel("Units")
plt.ylabel("Total Cost/Revenue")
plt.title("Revenue vs. Cost")
plt.show()
```

```python
plt.plot(xValues,yValues1)
plt.plot(xValues,yValues2)
plt.xlabel("Units")
plt.ylabel("Total Cost/Revenue")
plt.title("Revenue vs. Cost")
#We can call a legend by giving a list of the labels
plt.legend(["Cost","Revenue"])
plt.show()
```

```python
#Or we can use labels in the plot() function to declare it instead as well
plt.plot(xValues,yValues1,label="Cost")
plt.plot(xValues,yValues2,label="Revenue")
plt.xlabel("Units")
plt.ylabel("Total Cost/Revenue")
plt.title("Revenue vs. Cost")
plt.legend()
plt.show()
```

```python
#Color is an argument that let's us change the color of lines
plt.plot(xValues,yValues1,label="Cost",color="r")
plt.plot(xValues,yValues2,label="Revenue",color="g")

plt.xlabel("Units")
plt.ylabel("Total Cost/Revenue")
plt.title("Revenue vs. Cost")
plt.legend()
plt.show()
```

```python
yValues3 = [a-b for a,b in zip(yValues2,yValues1)]
print(yValues3)
```

```python
#Color is an argument that let's us change the color of lines
plt.plot(xValues,yValues1,label="Cost",color="r")
plt.plot(xValues,yValues2,label="Revenue",color="g")
plt.plot(xValues,yValues3,label="Profit",color="k")

plt.xlabel("Units")
plt.ylabel("Total Cost/Revenue")
plt.title("Revenue vs. Cost")
plt.legend()
plt.show()
```

```python
#Let's filter to get every 10th value in the list
print(xValues[::10])
print(yValues3[::10])
```

```python
plt.plot(xValues,yValues1,label="Cost",color="r")
plt.plot(xValues,yValues2,label="Revenue",color="g")
#We can also give a color code followed by a marker code as an argument. In this case k is black, o are the circle markers
#Plot the filtered values with the "ko" argument to get black circles
plt.plot(xValues[::10],yValues3[::10],"ko",label="Profit")

plt.xlabel("Units")
plt.ylabel("Total Cost/Revenue")
plt.title("Revenue vs. Cost")
plt.legend()
plt.show()
```
