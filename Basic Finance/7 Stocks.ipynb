```python
#Stocks are extremely hard to value
#In this course we will just start with the basics
#First, let's talk about dividends
```

```python
#What would we pay for a $5 dividend that we think we could get forever?
#This is easy, it's a perpetuity
r = .05
print(5/r)
```

```python
#What if we expect growth however, this complicates things a bit. 
#Let's look at a $5 dividend and say that it grows 2% each year, what is the nominal value each year
#2% growth works the same way as compound interest
t = list(range(26))
dividends1 = [5*(1.02)**x for x in t]
print(dividends1)
```

```python
import matplotlib.pyplot as plt
plt.plot(t,dividends1)
plt.xlabel("t")
plt.ylabel("Nominal Dividend Value")
plt.title("Nominal Dividend Value with 2% Growth")
plt.show()
```

```python
#For the present value of these dividends, we can apply our rate r to each one
dividends2 = [(5*(1.02)**x)/(1.05**x) for x in t]
print(dividends2)
```

```python
plt.plot(t,dividends1,label="Nominal")
plt.plot(t,dividends2,label="Present Value")
plt.legend()
plt.xlabel("t")
plt.ylabel("Nominal Dividend Value")
plt.title("Dividend Value with g=2% r=5%")
plt.show()
```

```python
#The equation for the value of a stock
print(5/(.05-.02))
#Growth drives value up since stocks are worth more in the future
```

```python
#Let's first experiment with different growth values and see how it affects dividends
g = [.02*x for x in range(0,6)]
r = .1
D0 = 5
data = []
for rate in g:
    #Get the dividend values over time
    dividends = [(D0*(1+rate)**x)/((1+r)**x) for x in t]
    data.append(dividends)
```

```python
for dividends,rate in zip(data,g):
    plt.plot(t,dividends,label="g="+str(int(rate*100))+"%")
plt.legend()
plt.xlabel("t")
plt.ylabel("Dividend Present Value")
plt.title("Dividend Value with r=10%")
plt.show()
```

```python
#You'll notice something odd when we try to apply the dividend discount model with growth > r
print(5/(.1-.11))
#Negative valuation? This is wrong, dividend discount model assumes that growth is not greater than r
#This asset would be worth infinity because we would always make more and more money adjusted for present value
```

```python
#We also get an error for growth equal to r
print(5/(.1-.1))
```

```python
#Let's see the valuation of stocks with different growth rates
r = .1
D0 = 5
for g in range(0,10):
    print("Growth Rate:",str(g)+"%","Present Value:",D0/(r-g/100))
```

```python
#We can get the dividend yield by dividing the dividend by the price of the stock
#For 0% growth rate:
print(5/50)
#For 5% growth rate:
print(5/100)
#Dividend yield goes down as we expect better growth in the dividends
```
