```python
dividend = 100
r = .1
year=0
years=[]
values=[]
for _ in range(101):
    years+=[year]
    values+=[dividend]
    year+=1
    dividend=dividend/(1+r)

import matplotlib.pyplot as plt
plt.plot(years, values)
plt.xlabel("Time")
plt.ylabel("Dividend Value")
plt.title("Constant Dividend Value over Time")
plt.show()
```

```python
import matplotlib.pyplot as plt
growth=0
for _ in range(6):
    dividend = 100
    r = .1
    year=0
    years=[]
    values=[]
    for _ in range(101):
        years+=[year]
        values+=[dividend]
        year+=1
        dividend=dividend*(1+growth)/(1+r)
    growth+=.02
    plt.plot(years, values)
plt.ylim((0,110))
plt.xlabel("Year")
plt.ylabel("Dividend Value")
plt.legend(["0%","2%","4%","6%","8%","10%"],title="Growth Rate")
plt.xlabel("Time")
plt.ylabel("Dividend Value")
plt.title("Varying Growth Rate Dividend Value over Time")
plt.show()
```

```python
def DD(div,g,r):
    return div/(r-g)
```

```python
print(DD(100,0,.1))
print(DD(100,.02,.1))
print(DD(100,.04,.1))
print(DD(100,.06,.1))
print(DD(100,.08,.1))
```
