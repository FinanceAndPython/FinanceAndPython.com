```python
#A perpetuity is a an infinite series of payments all of the same value
#It's value, however is not infinite because time value of money will eventually depreciate the value of payments to $0
```

```python
#For example, what is $100 worth after 1 year, 10 years, 100 years, 500 years, and 1000 years when r=2%
print(100/(1.02)**1)
print(100/(1.02)**10)
print(100/(1.02)**100)
print(100/(1.02)**500)
print(100/(1.02)**1000)
```

```python
#Let's graph it quickly as well
import matplotlib.pyplot as plt
t = list(range(1,1001))
PV = [100/((1.02)**x) for x in t]
plt.plot(t,PV)
plt.xlabel("t")
plt.ylabel("Payment PV")
plt.title("PV of $100 with r=2%")
plt.show()
```

```python
#Maybe we say that after $1000 years the rest of the payments have no effect
#Then our value would be the first 1000 years
print(sum(PV))
#Meaning we would pay $5000 dollars
```

```python
#We could get a rolling sum by period by using
#A libray itertools has a function accumulate which gives a rolling sum
from itertools import accumulate
totalPV = accumulate(PV)
print(totalPV)
#But we get back an object at first
```

```python
totalPV = list(totalPV)
#Convert it to a list to use it
print(totalPV)
```

```python
#Now we can also graph this
plt.plot(t,totalPV)
plt.xlabel("t")
plt.ylabel("Cummulative Present Value")
plt.title("Cummulative PV of $100 Perpetuity with r=2%")
plt.show()
```
