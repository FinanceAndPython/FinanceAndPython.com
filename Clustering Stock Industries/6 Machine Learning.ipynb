```python
import pandas as pd
df = pd.DataFrame.from_csv("RegressionMatrix1.csv",encoding="UTF-8")
df = df.fillna(0)
print(df)
```

```python
df = df.transpose()
print(df)
```

```python
for x in df.columns:
    df[x] = (df[x] - df[x].mean())/df[x].std(ddof=0)
print(df)
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=8)
```

```python
model = model.fit(df)
```

```python
print(model.cluster_centers_)
```

```python
print(model.labels_)
```

```python
df["Group"] = model.labels_
print(df)
```

```python
print(df["Group"]==1)
```

```python
print(df[df["Group"]==1])
```

```python
for x in range(8):
    print(df[df["Group"]==x])
```

```python
SP500 = pd.DataFrame.from_csv("SP500.csv", encoding="UTF-8")
tickers = []
for x in SP500["Ticker"].values:
    if "." in x:
        x = x.replace(".","")
    tickers.append(x)
SP500["Ticker"]=tickers
```

```python
SP500.index = SP500["Ticker"]
SP500 = pd.concat([SP500,df],axis=1,join="inner")
print(SP500)
```
