```python
countries = ["JPN","USA","SAU"]
```

```python
#Let's also set some indicators, we do this with a dictionary of world bank key to the name we want for it
indicators = {"NY.GDP.PETR.RT.ZS":"Oil Rents"}
indicators["NY.GDP.MKTP.CD"] = "GDP"
indicators["EN.ATM.CO2E.KT"] = "Pollution"
indicators["FP.CPI.TOTL.ZG"] = "Inflation"
```

```python
import datetime
dates = (datetime.datetime(1995, 1, 1), datetime.datetime(2017, 1, 1))
```

```python
df  = wbdata.get_dataframe(indicators, country=countries, data_date=dates)
print(df)
```

```python
print(df.index)
#This dataframe has multiple indexes! Both time and country
```

```python
#We can reset index to get rid of this multi-index
df.reset_index(inplace=True)
print(df)
```
