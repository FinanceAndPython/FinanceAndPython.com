```python
import pandas_datareader as pdr
#Pandas data-reader is the library we will use to get stock data and economic data from FRED
import datetime
#Datetime is the library for working with dates
start = datetime.datetime(2016, 1, 1)
#Dates are initialized with a year, month, then day as arguments
end = datetime.datetime(2017, 1, 1)
```

```python
df = pdr.get_data_yahoo("AAPL", start, end)
#Finally, the get_data_yahoo function takes a ticker, a start date, and an end date to get data
print(df)
#You may get a remote data error, this is an issue with Yahoo Finance, just try running the code again if this happens.
```

```python
print(df["Adj Close"])
#The adjusted close column gives adjusted close stock prices for each day
```

```python
df = pdr.get_data_yahoo(["AAPL","F"], start, end)
print(df)
#We are able to give a list of tickers to get multiple stocks
#The output isn't something we can see though since it is multiple dataframes
```

```python
df = df["Adj Close"]
#We can turn the dataframe into a 2d representation by indexing with "Adj Close", this will give the adjusted close for
#Each stock in our ticker list
print(df)
```

```python
#For whatever reason, the library turns the index upside down when multiple stocks are given
#We can fix it with sort index (but the operation does not happen in place)
print(df.sort_index())
```

```python
df = df.sort_index()
```

```python
print(df.pct_change())
#pct_change() is a function that gives us the percentage change from each row
```

```python
print(df.iloc[0])
#Let's see what our starting values are using iloc which returns the integer location items
```

```python
print(df/df.iloc[0])
#We can normalize by dividing each element of the column by its first value
#This allows us to see how much greater or less a stock is worth on a given day
```

```python
investment = [3000,4000]
print(df/df.iloc[0]*investment)
#If we multiplied this by an array representing the investment, we would be able to see the value of each investment
#At each point in time
```

```python
portfolio = df/df.iloc[0]*investment
#The wrong way to sum, it sums over the columns
print(portfolio.sum())
```

```python
#The right way, when we give the argument axis=1 we sum over the rows
print(portfolio.sum(axis=1))
```

```python
import matplotlib.pyplot as plt
#Matplotlib is a library for simple plotting
total = portfolio.sum(axis=1)
total.plot(kind="line")
#Using plot(kind="line") we create a matplotlib line plot
plt.show()
#We use plt.show() to show the plot
```

```python
portfolio.plot(kind="line")
plt.show()
#The plot() function plots each column, so in the case of portfolios we get both the Apple stock and the Ford stock plot
```

```python
portfolio["Total"] = portfolio.sum(axis=1)
portfolio.plot(kind="line")
plt.show()
```

```python
print(portfolio.index[0])
#The first day is January 4, 2016
```

```python
print(portfolio.index-portfolio.index[0])
#We can get the time delta or days since the first day by subtracting the start date from everything in the index
```

```python
delta = portfolio.index-portfolio.index[0]
print(delta.astype(int))
#If we change the type of delta to an integer, we get the seconds from the time delta
```

```python
#The numpy library has the ability to create a time delta measure
import numpy as np
print(np.timedelta64(1, 'Y'))
```

```python
print((delta)/np.timedelta64(1, 'Y'))
#This gives us the fraction of a year at each date
```

```python
time = (delta)/np.timedelta64(1, 'Y')
portfolio.index = time
#Change our index to equal the fraction of years past
portfolio.index.name = 't'
#And set the name of the index to t, since we don't have a date there anymore
print(portfolio)
```

```python
returns = portfolio/portfolio.iloc[0]
print(returns)
#Now let's get the absolute returns
```

```python
#Let's see how applying a function to each row works
def findReturn(row):
    print(row)
    print()
#The above function will take a row and then print it followed by a new line
portfolio.apply(findReturn,axis=1)
#We need to use axis=1 to work over rows instead of columns
```

```python
#Now let's change the function so that if we have the first row we return 0%, but otherwise we annualize the return
def findReturn(row):
    if row.name==0:
        return row-1
    else:
        return (row)**(1/row.name)-1
returns = returns.apply(findReturn,axis=1)
print(returns)
```

```python
print(returns.tail(100))
#Show only the last 100 rows
```
